```python
%cd /
from glob import glob

img_list_busnum = glob('/content/drive/MyDrive/BusNumTrainData/images/*')

print(len(img_list_busnum))
```

```python
from sklearn.model_selection import train_test_split

train_img_list_busnum, val_img_list_busnum = train_test_split(img_list_busnum, test_size=0.2, random_state=2000)

print(len(train_img_list_busnum), len(val_img_list_busnum))
```

```python
with open('/content/drive/MyDrive/BusNumTrainData/train.txt', 'w') as f:
  f.write('\n'.join(train_img_list_busnum))

with open('/content/drive/MyDrive/BusNumTrainData/val.txt', 'w') as f:
  f.write('\n'.join(val_img_list_busnum))
```

```python
import yaml

with open('/content/drive/MyDrive/BusNumTrainData/data.yaml', 'r') as f:
  data = yaml.load(f)
print(data)

data['train'] = '/content/drive/MyDrive/BusNumTrainData/train.txt'
data['val'] = '/content/drive/MyDrive/BusNumTrainData/val.txt'

with open('/content/drive/MyDrive/BusNumTrainData/data.yaml', 'w') as f:
  yaml.dump(data, f)
print(data)
```
